```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots

pd.options.mode.chained_assignment = None

# Read in the data
df = pd.read_csv(r'movies.csv')
```

```python
df
```

```python
# Checking for any missing data by looping through the data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for columns
print(df.dtypes)
```

```python
# Are there any Outliers?
df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
```

```python
# Order data
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Scatter plot with gross vs budget
sns.regplot(x="gross", y="budget", data=df)
plt.title("gross vs budget")
```

```python
# Scatter plot with score vs gross
sns.regplot(x="score", y="gross", data=df)
plt.title("score vs gross")
```

```python
# Correlation Matrix between all numeric columns
df.corr(method='pearson', numeric_only=True)

```

```python
correlation_matrix = df.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# High correlation (> 0.5)
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

with pd.option_context('display.max_rows', None):
    print(strong_pairs)
```
